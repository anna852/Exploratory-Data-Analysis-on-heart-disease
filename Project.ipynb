We have seen from some of the plots above that there is a difference between the percentage of  males patients with heart disease and the percentage of females patients with heart disease- females are more likely to get heart disease.<br>
Do females really have heart disease more than males?

We will use a bootstrap method to compute a 95% confidence interval for the difference between the sick female rate and the sick male rate.

- Based on the article from NCBI (https://www.ncbi.nlm.nih.gov/pmc/articles/PMC5435266/) coronary heart disease mortality was higher in males than in females throughout adulthood. (but the magnitude of the difference varied by age. Men-to-women CHD mortality rate ratios were 4–5 in middle age (30–64 years) and 2 thereafter (65–89 years)dataset). Does it mean that males are also sicker than females?

So we wanted to explore this question, and based on second question graphs we assume that heart disease rate is higher at females.<br>
picture was taking from https://health.clevelandclinic.org/women-men-higher-risk-heart-attack/.

H0:In the population, there is no difference between male and female heart disease rate.<br>
H1: In the population, female heart disease rate is different (higher) than males.

We will test our zero hypothesis by examining the difference between the percentage of sick females out of all females and the percentage of sick males out of all males - whether the difference is equal to 0 (whether 0 is in a confidence interval of 95%).

Since patient / non-patient is distinguished by 1 and 0 respectively, we can actually calculate the average of the sick females and the sick males and find the difference between them.

```python
# function that returns the difference in averages of "column_name" grouping by "grouping_var".
# This is how we will calculate the difference between distributions.

def diff_of_avgs(df, column_name, grouping_var):
    grpby_var = df.groupby(grouping_var)
    avgs = grpby_var[column_name].mean()
    return avgs[0] - avgs[1]

# # We wanted to make sure that indeed the function of the mean would give us the desired answer,
# # so we wrote ourselves a function that would calculate the percentage difference.
# # Writing the function and thinking about it took a long time,
# # so, we would be happy to keep it even though it expresses exactly the same result as in the previous function.

# def func (df,grouping_var):
#     gb_sex = df.groupby(grouping_var)
#     gb_fm=gb_sex.get_group('Female')
#     gb_m=gb_sex.get_group('Male')
#     precentage_fm=len(gb_fm[gb_fm.target == 1])/len(gb_fm)
#     precentage_m=len(gb_m[gb_m.target == 1])/len(gb_m)
#     return precentage_fm-precentage_m


def bootstrap_mean_difference(original_sample, column_name, grouping_var, num_replications):
    '''This function returns an array of bootstrapped differences between two sample averages:
      original_sample: df containing the original sample
      column_name: name of column containing the variable to average
      grouping_var: name of variable according to which to group
      num_replications: number of bootstrap samples'''
    original_sample_size = original_sample.shape[0] # we need to replicate with the same sample size
    original_sample_cols_of_interest = original_sample[[column_name, grouping_var]]
    bstrap_mean_diffs = np.empty(num_replications)# same size as nimber of replications
    for i in range(num_replications):
        bootstrap_sample = original_sample_cols_of_interest.sample(original_sample_size, replace=True) # note WITH REPLACEMENT!
        resampled_mean_diff = diff_of_avgs(bootstrap_sample, column_name, grouping_var)# calculating the difference
        bstrap_mean_diffs[i] = resampled_mean_diff
    
    return bstrap_mean_diffs

# run the bootstrap procedure
bstrap_diffs = bootstrap_mean_difference(heart_df, 'target', 'sex', 7000)# Run 7000 simulations to get more accuracy data.

# Get the endpoints of the 95% confidence interval
left_end = np.percentile(bstrap_diffs, 2.5, interpolation='higher')
right_end = np.percentile(bstrap_diffs, 97.5, interpolation='higher')
print('The 95% boostsrap confidence interval for difference between population means', [left_end,right_end])

# visualize results
ax = sns.displot(bstrap_diffs)
plt.hlines(y=0, xmin=left_end, xmax=right_end, colors='orange', linestyles='solid', lw=7, clip_on=False);  
ax.set(xlabel='Difference between rate of heart disease', ylabel='frequency',
       title='Distribution of bootstrap estimates for the difference between\nheart disease rate of males and females');

plt.scatter(0, 0, color='red', s=120, clip_on=False)
```

We set our confidence interval to be 95%. It can be seen that 0 is not included in our confidence interval (0 represents that there is no difference between morbidity rates in women and men) and therefore the null hypothesis can be rejected.

The conclusion is that there is a difference between the  sick females rates and the sick males rates with heart disease.

**Reminder, we have no missing values.**

```python
# we need to transform our categorical variables to dummies,Even those classified as numeric. 
knn_df = pd.get_dummies(knn_df, columns=['exang','sex','cp'],
                       drop_first=True) # we should drop the redundant category
knn_df.head()
```

- We choose to divide our data into 2, where the train will be 80% and the test will be 20%. It is to use the knn method for prediction.

```python
# we need to split our data to train and test sets
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split

# It is always sensible to randomly shuffle the dataframe first
knn_df = knn_df.sample(frac=1)

# Split to X and Y
X = knn_df.loc[:, knn_df.columns != 'target'] # features
Y = knn_df.loc[:, 'target'].values # labels

# Split to train and test
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.20)
X_train
```

```python
# We need to scale the variables to be on the same scale
# We choose here to standardize using MinMaxScaler
from sklearn.preprocessing import MinMaxScaler
df_columns = X_train.columns
scaler = MinMaxScaler()
scaled_X_train = scaler.fit_transform(X_train)
# we will do a scaling only on the training set, and then we will transform the test set to the resulting scale
scaled_X_test = scaler.transform(X_test)

scaled_df = pd.DataFrame(scaled_X_train, columns=df_columns)
scaled_df.describe()
```

```python
# Finally, we can run 5-fold CV to find optimal k - using only the training data
from sklearn.model_selection import cross_val_score

mean_cv_scores = []
k_list = range(1, 20)
for nn in k_list:
    knn_cv = KNeighborsClassifier(n_neighbors=nn)
    cv_scores = cross_val_score(knn_cv, scaled_X_train, Y_train, cv=5)
    mean_cv_scores.append(cv_scores.mean())
    
# output results
best_k = mean_cv_scores.index(max(mean_cv_scores))+1 # gets index of best performing k and adds 1
print('Highest accuracy is obtained for k =', best_k)
plt.plot(k_list, mean_cv_scores, '-o')
plt.xlabel('Number of neighbors (k)')
plt.ylabel('Accuracy');
```

```python
# Last, we need to retrain our chosen kNN on the whole training data and test its accuracy on the test data
from sklearn.metrics import confusion_matrix
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score

knn_classifier = KNeighborsClassifier(n_neighbors=best_k)  
knn_classifier.fit(scaled_X_train, Y_train)
print('accuracy of the classifier is', knn_classifier.score(scaled_X_test, Y_test))

# Compute a confusion matrix
predictions = knn_classifier.predict(X=scaled_X_test) # get the classifier's predictions 
print('confusion matrix: \n', confusion_matrix(y_true=Y_test, y_pred=predictions, labels=[0, 1])) # rows are true values, columns are predicted values

print('precision: ', precision_score(y_true=Y_test, y_pred=predictions, labels=[0, 1]))
print('recall: ', recall_score(y_true=Y_test, y_pred=predictions, labels=[0, 1]))
```

```python
from sklearn.metrics import plot_confusion_matrix
plot_confusion_matrix(knn_classifier, X_test, Y_test,
                     cmap=plt.cm.Reds)
```
